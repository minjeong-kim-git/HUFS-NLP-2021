# 수많은 데이터에서 특정 조건을 만족하는 데이터만 다루기
타이타닉 탑승자 중 나이가 35세 이상인 사람만 필요하다고 하면 아래와 같이 코드를 작성할 수 있습니다.

```titanic["Age"] > 35```은 titanic의 요소 중 Age가 35보다 크다면 1(True)를, 작거나 같다면 0(False)를 반환하며 titanic의 모든 row를 탐색합니다.




````
titanic["Age"] > 35
````

```titanic["Age"] > 35```은 titanic[*]에서 *의 결과가 1(True)인 row만 출력하게 됩니다.



````
above_35 = titanic[titanic["Age"] > 35]
above_35
````

````
above_35.shape
````

이번에는 2 또는 3등급 숙소를 사용하는 탑승객을 찾아봅시다.

```titanic["Pclass"].isin([2, 3])```은 titanic에서 Pclass 요소를 확인하여 2 혹은 3인 경우에만 1(True), 아니라면 0(False)를 반환하도록 합니다.



````
class_23 = titanic[titanic["Pclass"].isin([2, 3])]
class_23
````

기본적인 Python 논리 연산을 통한 계산도 가능합니다.

````
class_23 = titanic[(titanic["Pclass"] == 2) | (titanic["Pclass"] == 3)]
class_23
````

수많은 데이터들 중에서는 오류가 있거나 확인되지 않음을 이유로 빈칸이나 NaN(이용불가) 값을 가진 경우가 있습니다.

이러한 값을 제외하려면 아래와 같은 코드를 사용합니다.

````
age_no_na = titanic[titanic["Age"].notna()]
age_no_na
````

이름에 특정 단어가 들어가는 사람의 데이터만 얻으려면 아래와 같이 코드를 작성합니다.

````
titanic["Name"].str.contains("Countess")
````

````
titanic[titanic["Name"].str.contains("Countess")]
````
